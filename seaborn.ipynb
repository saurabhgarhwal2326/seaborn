```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print('libraries imported')
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
# sns.lineplot()
# sns.barplot()
# sns.boxplot()
# sns.scatterplot()
# sns.histplot
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.show()
```

```python
sns.relplot(y='tip',x='total_bill',data=df,hue="sex")

plt.show()
```

```python
sns.relplot(y='tip',x='total_bill',data=df,hue="sex",style='time')
plt.show()
```

```python
sns.relplot(y='tip',x='total_bill',data=df,hue="time",style='sex',col='day')
```
